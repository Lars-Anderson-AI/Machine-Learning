```python
import pandas as pd
home_data = pd.read_csv('NY-House-Dataset.csv', usecols = ['BEDS', 'BATH', 'PRICE'])
home_data.head()
```

```python
import seaborn as sns
sns.scatterplot(data = home_data, x = 'BEDS', y = 'BATH', hue = 'PRICE')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(home_data[['BEDS', 'BATH']], home_data[['PRICE']], test_size = 0.33, random_state = 0)
```

```python
from sklearn import preprocessing
x_train_norm = preprocessing.normalize(x_train)
x_test_norm = preprocessing.normalize(x_test)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(x_train_norm)
```
